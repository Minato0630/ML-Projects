```python
import matplotlib.pyplot as plt
import numpy as np
import warnings
import pickle
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay

warnings.simplefilter("ignore", DeprecationWarning)
data = load_iris()
X, y = data.data, data.target
def train_evaluate(test_size=0.2):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42)
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    model = LogisticRegression(max_iter=500)
    model.fit(X_train_scaled, y_train)
    y_pred = model.predict(X_test_scaled)
    accuracy = model.score(X_test_scaled, y_test)
    
    print(f"\n--- Test Size: {int(test_size*100)}% ---")
    print("Model Accuracy:", accuracy)
    
    # Extra Credit 1: Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    print("\nConfusion Matrix:")
    print(cm)
    ConfusionMatrixDisplay(cm, display_labels=data.target_names).plot(cmap=plt.cm.Blues)
    plt.show()
    
    # Extra Credit 2: Classification Report
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred, target_names=data.target_names))
    
    # Extra Credit 3: Feature Importance using model coefficients
    feature_importance = model.coef_
    plt.figure(figsize=(8,5))
    for i, class_name in enumerate(data.target_names):
        plt.bar(np.arange(X.shape[1]) + i*0.2, feature_importance[i], width=0.2, label=class_name)
    plt.xticks(np.arange(X.shape[1]), data.feature_names, rotation=45)
    plt.ylabel("Coefficient Value")
    plt.title("Feature Importance (Model Coefficients)")
    plt.legend()
    plt.show()
    
    return model, scaler

# Extra Credit 4: Experiment with different test sizes
models = {}
for ts in [0.2, 0.3, 0.4]:
    model, scaler = train_evaluate(test_size=ts)
    models[ts] = (model, scaler)

# Extra Credit 5: Save the model and scaler
with open("iris_model.pkl", "wb") as f:
    pickle.dump(models[0.2][0], f)
with open("iris_scaler.pkl", "wb") as f:
    pickle.dump(models[0.2][1], f)

# Load the model and make a prediction
with open("iris_model.pkl", "rb") as f:
    loaded_model = pickle.load(f)
with open("iris_scaler.pkl", "rb") as f:
    loaded_scaler = pickle.load(f)

# Example prediction
sample = np.array([[5.1, 3.5, 1.4, 0.2]])  # Example iris features
sample_scaled = loaded_scaler.transform(sample)
pred_class = loaded_model.predict(sample_scaled)
print("\nPrediction for sample input:", data.target_names[pred_class[0]])
```
