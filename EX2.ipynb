```python
import matplotlib.pyplot as plt 
import numpy as np 
from sklearn.linear_model import LinearRegression 
from sklearn.linear_model import SGDRegressor 
x = np.array([1,2,4,3,5]) 
y = np.array([1,3,4,3,7]) 
print('\n',x.shape) 
print('\n',y.shape) 
print('\n',x) 
print('\n',y)
x = x.reshape(-1,1) 
print('\n',x.shape) 
print('\n',x)
model = LinearRegression() 
model.fit(x, y) 
print(model.predict([[8]])) 
y_pred = model.predict(x) 
plt.scatter(x, y, color='r', marker='o') 
plt.plot(x, y_pred, color='b', linewidth=2) 
plt.xlabel('x') 
plt.ylabel('y') 
plt.show() 
print('\n',model.coef_) 
print('\n',model.intercept_)
from sklearn.preprocessing import StandardScaler 
scaler = StandardScaler() 
x_stand = scaler.fit_transform(x.astype(float)) 
sgdr = SGDRegressor(penalty='l2', alpha=0.15, max_iter=100)
sgdr.fit(x_stand, y) 
print(sgdr.predict([[10]]))
y_pred2 = sgdr.predict(x_stand) 
plt.scatter(x_stand, y, color='g', marker='o') 
plt.plot(x_stand, y_pred2, color='b', linewidth=2) 
plt.xlabel('x_normalized') 
plt.ylabel('y') 
plt.show() 
print(sgdr.coef_) 
print(sgdr.intercept_)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import SGDRegressor

# 1. Read your dataset
data = pd.read_csv("car_data.csv")

print("\nDataset Head:\n", data.head())

# Use Mileage as feature and Price as target
x = data['Mileage'].values
y = data['Price'].values

print('\nShape of x:', x.shape)
print('Shape of y:', y.shape)
print('\nx sample:', x[:10])
print('y sample:', y[:10])

# reshape x
x = x.reshape(-1, 1)
print('\nReshaped x:', x.shape)
print(x[:10])

# 2. Linear Regression Model
model = LinearRegression()
model.fit(x, y)

y_pred = model.predict(x)

# scatter + line
plt.scatter(x, y, color='r', marker='o')
plt.plot(x, y_pred, color='b', linewidth=2)
plt.xlabel('Mileage')
plt.ylabel('Price')
plt.title('Linear Regression - Car Price Prediction')
plt.show()

# 3. Print coefficient & intercept
print("\nLinear Regression Coefficient:", model.coef_)
print("Linear Regression Intercept:", model.intercept_)

# 4. SGD Regressor Model
sgdr = SGDRegressor(max_iter=100)
sgdr.fit(x, y)

y_pred2 = sgdr.predict(x)

# scatter + line
plt.scatter(x, y, color='g', marker='o')
plt.plot(x, y_pred2, color='b', linewidth=2)
plt.xlabel('Mileage')
plt.ylabel('Price')
plt.title('SGD Regression - Car Price Prediction')
plt.show()

# 5. Print SGD weights
print("\nSGD Coefficient:", sgdr.coef_)
print("SGD Intercept:", sgdr.intercept_)
```
